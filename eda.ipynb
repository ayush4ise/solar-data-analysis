- Collecting the data for Jakarta from 1981-2023 and saving it in a big dataframe
- Not considering the data from 1980 as a few intial values are missing

````python
# Jakarta 

data_jakarta = pd.read_csv('1980-2023 renewable energy data/ninja_pv_-7.2623_112.7361_1980.csv', header=3)
name = "1980-2023 renewable energy data/ninja_pv_-7.2623_112.7361_"

for i in range(1981, 2024):
    data_jakarta = pd.concat([data_jakarta, pd.read_csv(name + str(i) + '.csv', header=3)])

data_jakarta.drop(columns=['time'], inplace=True)
data_jakarta['local_time'] = pd.to_datetime(data_jakarta['local_time'])

print("Info:\n", data_jakarta.info())
print("Head:\n", data_jakarta.head())
print("Tail:\n", data_jakarta.tail())
````

## Average for Every Month for Every Year

- Converting the data to calculate the average kW for every month for every year.

````python
monthly_means = {}

for month in range(1,13):
    for hour in range(24):
        monthly_means[(month, hour)] = [data_jakarta[(data_jakarta['local_time'].dt.year == year) & (data_jakarta['local_time'].dt.month == month) & (data_jakarta['local_time'].dt.hour == hour)]['electricity'].mean() for year in range(1981, 2024)]

monthly_means = pd.DataFrame(monthly_means, index=range(1981, 2024))

print("Monthly means:\n", monthly_means)
````

- Taking the transformed data and saving it to an Excel file for further analysis

````python
month_to_name = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June', 7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'}

monthly_means.xs(1, level=0, axis=1).to_excel('transformed_data/monthly_means_jakarta.xlsx', sheet_name='January')
for month in range(2, 13):
    # write to the same excel file
    with pd.ExcelWriter('transformed_data/monthly_means_jakarta.xlsx', engine='openpyxl', mode='a') as writer:
        monthly_means.xs(month, level=0, axis=1).to_excel(writer, sheet_name=month_to_name[month])
````

- Making six sigma plots for every hour of every month and finding outliers
- Storing the visualizations in ```visualizations``` folder

````python
import numpy as np
import matplotlib.pyplot as plt

def make_sigma_plot(data, col, month):
    mean = data[col].mean()
    std = data[col].std()

    data[col].plot()
    plt.axhline(mean, color='r')
    plt.axhline(mean + 3*std, color='g')
    plt.axhline(mean - 3*std, color='g')

    outliers = data[col][(data[col] > mean + 3*std) | (data[col] < mean - 3*std)]   

    for i in outliers.index:
        plt.text(i, outliers[i], str(i))

    plt.title(f'{col}th hour')
    plt.xlabel('Year')
    plt.ylabel('Electricity (in kW)')
    plt.savefig(f'visualizations/sigma_plots/{month}_{col}th_hour.png')
    plt.close()


sheet_names = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

for sheet in sheet_names:
    data = pd.read_excel('transformed_data/monthly_means_jakarta.xlsx', sheet_name=sheet, index_col=0)
    for col in data.columns:
        make_sigma_plot(data, col, sheet)
````

## Box Plots

- Making box plots for every month
- Storing the visualizations in ```visualizations``` folder 

````python
def make_box_plot(data, month):
    plt.figure(figsize=(10, 6))
    data.boxplot()
    plt.title(f'Electricity production by hour {month}')
    plt.ylabel('Electricity (in kW)')
    plt.savefig(f'visualizations/box_plots/{month}.png')
    plt.close()

for sheet in sheet_names:
    data = pd.read_excel('transformed_data/monthly_means_jakarta.xlsx', sheet_name=sheet, index_col=0)
    make_box_plot(data, sheet)
````

## ANOVA

- Performing ANOVA on the data to check if the means of the data (for a given hour and given month) are different for different years
- Storing the results in ```results``` folder

````python
from scipy.stats import f_oneway

def anova_test(data):
    f, p = f_oneway(*data)
    return f, p

data_anova = {}

for month in range(1, 13):
    data_anova[month] = []
    for hour in range(24):
        data = [list(data_jakarta[(data_jakarta['local_time'].dt.year == year) & (data_jakarta['local_time'].dt.month == month) & (data_jakarta['local_time'].dt.hour == hour)]['electricity']) for year in range(1981, 2024)]
        f, p = anova_test(data)
        data_anova[month].append([f, p])

data_anova = pd.DataFrame(data_anova, index=range(24))

data_anova.head()
````
